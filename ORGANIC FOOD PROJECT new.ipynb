```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
brand_name=[]
price_rs=[]
rating_s = []
review_s=[]
original_price=[]
quantity_s=[]
varitie_s=[]
discount_s=[]
page_num=input("Enter number of pages")
for i in range(1,int(page_num)+1):
    url="https://www.flipkart.com/search?q=organic+food+&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="+str(i)
    req=requests.get(url)
    content=BeautifulSoup(req.content,"html.parser")
    brand=content.find_all("a",{"class":"s1Q9rs"})
    price=content.find_all("div",{"class":"_30jeq3"})
    rating=content.find_all("div",{"class":"_3LWZlK"})
    review=content.find_all("span",{"class":"_2_R_DZ"})
    originalprice=content.find_all("div",{"class":"_3I9_wc"})
    quantity=content.find_all("div",{"class":"_3Djpdu"})
    varities=content.find_all("a",{"class":"s1Q9rs"})
    discount=content.find_all("div",{"class":"_3Ay6Sb"})
    print("organic in page"+str(i))
    print(len(brand))
    
    
    for i in brand:
        brand_name.append(i.text.split()[0])
    for i in price:
        price_rs.append(i.text)
    for i in rating:
        rating_s.append(i.text)
    for i in review:
        review_s.append(i.text.strip("()"))
    for i in originalprice:
        original_price.append(i.text)
    for i in quantity:
        quantity_s.append(i.text)
    for i in varities:
        varitie_s.append(i.text.split(",")[0].split()[1])
    for i in discount:
        discount_s.append(i.text.strip("% off"))
```
